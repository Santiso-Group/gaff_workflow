```python
import mbuild as mb

#Import molecule from file

class MeOH(mb.Compound):
    def __init__(self):
        super(MeOH, self).__init__()
        
        mb.load('methanol.mol2', compound=self)
        
methanol = MeOH()
methanol.visualize()
```

```python
# Calculating molecule charges using foyer

import foyer

# For this we need to import antefoyer
import antefoyer

# Load GAFF forcefild 
gaff = foyer.forcefields.load_GAFF()

# Identify GAFF atomtypes with antechamber
typed_molecule = antefoyer.ante_atomtyping(methanol, 'gaff')

# Use foyer to parameterize
parameterized_molecule = gaff.parametrize_system(typed_molecule)

# We apply AM1-BCC charges from antechamber
molecule_with_charges = antefoyer.ante_charges(parameterized_molecule,'bcc')

# Save to whatever file format you desire
molecule_with_charges.save('methanol_with_charges.mol2', overwrite=True)
```

```python
#Creating a solvent box

#Dimensions of the box
box = mb.Box(lengths=[3, 3, 3])

#Number of molecules in the box
box_of_methanol_3 = mb.fill_box(compound=methanol, n_compounds=402, box=box)
box_of_methanol_3.visualize()
```

```python
#AtomTyping

#ethane_box = mb.fill_box(compound=Ethane(), n_compounds=100, box=[2, 2, 2])
box_of_methanol_3.save('box_of_methanol_3.xyz', overwrite=True)
box_of_methanol_3.save('box_of_methanol_3.pdb', overwrite=True)
box_of_methanol_3.save('box_of_methanol_3.lammps', forcefield_name='gaff', overwrite=True)
```
